DATA CLEANING

Check for Null Values in Sale Data

```python
print("\n--- Null Values in Sales Data ---")
print(df_sales.isnull().sum())
```

Handle Null Values in 'City' and 'Delivery Status' by dropping Rows

```python
df_sales.dropna(subset=["city", "delivery_status"], inplace=True)
print("\nNulls after dropping rows in city/delivery_status:")
print(df_sales[["city", "delivery_status"]].isnull().sum())
```

Handle Null Values in 'Delivery Time' by filling with the mean

```python
mean_delivery_time = df_sales["delivery_time_mins"].mean()
df_sales["delivery_time_mins"].fillna(mean_delivery_time, inplace=True)
print("\nNulls after filling mean in delivery_time_mins:")
print(df_sales["delivery_time_mins"].isnull().sum())
```

Check fro Duplicate Records

```python
print("\n--- Duplicate Records in Sales Data ---")
print(f"Number of duplicate rows: {df_sales.duplicated().sum()}")
```

Remove Duplicate Records

```python
df_sales.drop_duplicates(inplace=True)
print(f"Number of rows after removing duplicates: {df_sales.shape[0]}")
```

Convert 'Order Date' to Datetime Objects

```python
df_sales["order_date"] = pd.to_datetime(df_sales["order_date"])
```

Data Analysis - Aggregations and Grouping

Find min, max and average total amount

```python
min_amount = df_sales["total_amount"].min()
max_amount = df_sales["total_amount"].max()
avg_amount = df_sales["total_amount"].mean()
print(f"\nMin Total Amount: {min_amount:.2f}")
print(f"Max Total Amount: {max_amount:.2f}")
print(f"Average Total Amount: {avg_amount:.2f}")
```

Top 5 Products by Total Sales Amount

```python
top_products = (
    df_sales
    .groupby("product_id")["total_amount"]
    .sum()
    .nlargest(5)
)


print("\n--- Top 5 Products by Sales Amount ---")
print(top_products)

```

Merge with Product Details to get Product names

```python
top_products_details = top_products.reset_index().merge(df_products,
on="product_id")
print("\n--- Top 5 Products by Sales Amount (with names) ---")
print(top_products_details[["product_name", "category", "total_amount"]])
```

Total Sales by City

```python
sales_by_city = (
    df_sales
    .groupby("city")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)


print("\n--- Total Sales by City ---")
print(sales_by_city)

```

Average Delivery Time by City

```python
avg_delivery_time_by_city = (
    df_sales
    .groupby("city")["delivery_time_mins"]
    .mean()
    .sort_values()
)


print("\n--- Average Delivery Time by City (minutes) ---")
print(avg_delivery_time_by_city)

```

Sale Trend over Time

```python
df_sales["month"] = df_sales["order_date"].dt.to_period("M")
monthly_sales = df_sales.groupby("month")["total_amount"].sum()
print("\n--- Monthly Sales Trend ---")
print(monthly_sales)
```

Sale by Product Category

```python
sales_by_category = df_sales.merge(df_products, on="product_id")
sales_by_category = (
    sales_by_category
    .groupby("category")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)


print("\n--- Total Sales by Product Category ---")
print(sales_by_category)

```

DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

fig, axes = plt.subplots(3, 3, figsize=(22, 18))
fig.suptitle("Sales Analysis Dashboard", fontsize=22)

# 1 Top 5 Products
sns.barplot(
    x="product_name",
    y="total_amount",
    data=top_products_details,
    palette="viridis",
    ax=axes[0, 0]
)
axes[0, 0].set_title("Top 5 Products by Sales")
axes[0, 0].tick_params(axis='x', rotation=45)

# 2 Sales by City
sns.barplot(
    x=sales_by_city.index,
    y=sales_by_city.values,
    palette="magma",
    ax=axes[0, 1]
)
axes[0, 1].set_title("Total Sales by City")
axes[0, 1].tick_params(axis='x', rotation=45)

# 3 Monthly Sales Trend
monthly_sales.index = monthly_sales.index.to_timestamp()
sns.lineplot(
    x=monthly_sales.index,
    y=monthly_sales.values,
    marker="o",
    ax=axes[0, 2]
)
axes[0, 2].set_title("Monthly Sales Trend")
axes[0, 2].tick_params(axis='x', rotation=45)

# 4 Sales by Category
sns.barplot(
    x=sales_by_category.index,
    y=sales_by_category.values,
    palette="pastel",
    ax=axes[1, 0]
)
axes[1, 0].set_title("Sales by Category")
axes[1, 0].tick_params(axis='x', rotation=45)

# 5 Delivery Time Distribution
sns.histplot(
    df_sales["delivery_time_mins"],
    bins=20,
    kde=True,
    color="#2ECC71",
    ax=axes[1, 1]
)
axes[1, 1].set_title("Delivery Time Distribution")

# 6 Order Status Count
sns.countplot(
    x="delivery_status",
    data=df_sales,
    palette="Set2",
    ax=axes[1, 2]
)
axes[1, 2].set_title("Order Status Count")
axes[1, 2].tick_params(axis='x', rotation=45)

# 7 Payment Method Distribution
sns.countplot(
    x="delivery_status",
    data=df_sales,
    palette="coolwarm",
    ax=axes[2, 0]
)
axes[2, 0].set_title("Payment Method Distribution")
axes[2, 0].tick_params(axis='x', rotation=45)

# 8 Average Delivery Time by City
sns.barplot(
    x="city",
    y="delivery_time_mins",
    data=df_sales,
    estimator=sum,
    palette="cubehelix",
    ax=axes[2, 1]
)
axes[2, 1].set_title("Total Delivery Time by City")
axes[2, 1].tick_params(axis='x', rotation=45)

# 9 Sales vs Quantity Scatter
sns.scatterplot(
    x="quantity",
    y="total_amount",
    data=df_sales,
    hue="delivery_status",
    ax=axes[2, 2]
)
axes[2, 2].set_title("Sales vs Quantity")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
